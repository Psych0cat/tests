### предобработка текста

```python
with open('text.txt') as f:
    content = f.readlines()
```

```python
text = ' '.join(content)
```

```python
import re
regex = re.compile("[А-Яа-я----]+")

def words_only(text, regex=regex):
    return " ".join(regex.findall(text))
```

```python
text = words_only(text, regex)
```

```python
text = text.lower()
```

```python
from pymystem3 import Mystem
```

```python
m = Mystem()
```

```python
def lemmatize(text, mystem=m):
    try:
        return "".join(m.lemmatize(text)).strip()  
    except:
        return " "
```

```python
text = lemmatize(text)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
mystopwords = stopwords.words('russian') + ['это', 'наш' , 'тыс', 'млн', 'млрд', 'также',  'т', 'д', '-', '-']
```

```python
def  remove_stopwords(text, mystopwords = mystopwords):
    try:
        return " ".join([token for token in text.split() if not token in mystopwords])
    except:
        return ""
```

```python
text = remove_stopwords(text)
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
```

```python
def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]
```

###  Нарезать текст на биграммы

```python
trigrams = get_ngrams(text, 3)
```

```python
bigrams = get_ngrams(text, 2)
```

```python
len(bigrams)
```

```python
len(trigrams)
```

```python
from collections import Counter, OrderedDict
```

###  Сколько раз встречались биграммы и триграммы

```python
dict_bigrams = dict(Counter(bigrams))
```

```python
dict_bigrams = OrderedDict(dict_bigrams)
```

```python
dict_trigrams = dict(Counter(trigrams))
```

```python
dict_trigrams = OrderedDict(dict_trigrams)
```

### Вероятности биграмм

```python
p_bigrams={}
for key, value in dict(Counter(bigrams)).items():
    p_bigrams[key] = value/1734
```

```python
p_bigrams = OrderedDict(p_bigrams)
```

### Вероятности триграмм

```python
p_trigrams={}
for key, value in dict(Counter(trigrams)).items():
    p_trigrams[key] = value/1735
```

```python
p_trigrams = OrderedDict(p_trigrams)
```

 ### Триграммы, содержащие биграммы

```python
lst_two_elements = []
for i in range(len(trigrams)):
    lst_two_elements.append(trigrams[i].split()[0:2])
```

```python
strings_to_compare = []
for i in range(len(lst_two_elements)):
    strings_to_compare.append(' '.join(lst_two_elements[i]))
```

```python
trigrams_including_bigrams = []
```

```python
for i in range(len(strings_to_compare)):
    if strings_to_compare[i] in bigrams:
        trigrams_including_bigrams.append(strings_to_compare[i])
    else: trigrams_including_bigrams.append(0)
```

###  Посчитать вероятность триграммы  как - число встреч триграммы / (число похожих биграмм).

```python
[x/y for x, y in zip(list(dict_trigrams.values()), list(Counter(trigrams_including_bigrams).values()))]
```

### Вывод общей информации

```python
for i in range(len(bigrams)):
    print(f'биграмма "{bigrams[i]}" встречется {list(dict_bigrams.values())[i]} раз с вероятностью {list(p_bigrams.values())[i]}')
```

```python
for i in range(len(trigrams)):
    print(f'триграмма "{trigrams[i]}" встречется {list(dict_trigrams.values())[i]} раз с вероятностью {list(p_trigrams.values())[i]}')
```
